```python
x = np.arange(2, 5, 1) 
y = 2**(x*(x-1)/2) 
z = np.cumprod(x)
plt.title('y=|x|') 
plt.xlabel('x')    
plt.ylabel('y')    
plt.plot(x, y,label='y')  
plt.plot(x, z,label='z')  
plt.legend()
plt.show() 
```
